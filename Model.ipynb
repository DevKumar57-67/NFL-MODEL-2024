```python
#Using Linear Regression on the dataset
x = nfl[["homeFinalScore"]]
y = nfl[["visitorFinalScore"]]
```

```python
#Splitting the dataset
x_train,x_test,y_train,y_test = train_test_split (x,y,test_size=0.5)
```

```python
#Importing Linear Regression Algorithm
from sklearn.linear_model import LinearRegression
```

```python
#Declaring  Linear Regression
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
y_pred=lr.predict(x_test)
```

```python
#Calculating the Error in the Model
print("The model is Accurate ")
mean_squared_error(y_train,y_pred)
```
